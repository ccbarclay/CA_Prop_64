```python
import pandas as pd
```

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
props.info()
```

```python
contribs.info()
```

```python
props.prop_name.value_counts()
```

```python
prop63 = props[props.prop_name == 'PROPOSITION 063- FIREARMS. AMMUNITION SALES. INTIATIVE STATUTE.']
```

```python
prop63.head()
```

```python
merged = pd.merge(prop63, contribs, on="calaccess_committee_id").sort_values("amount", ascending=False)
```

```python
merged.info()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == 'SUPPORT']
```

```python
oppose = merged[merged.committee_position == 'OPPOSE']
```

```python
support.head()
```

```python
oppose.head()
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
merged.groupby(["committee_name_x", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
oppose[oppose.amount > 1000].sort_values("amount", ascending=False)
```
